## Read/Show a row (using 'integer location', i.e. iloc[])

```python
first_row = df.iloc[0]
first_row
```

## Read in multiple rows

```python
multi_rows = df.iloc[[1, 3, 5]]
multi_rows
```

---

## Read in multiple rows with select columns (using iloc, i.e integer locations of both rows and locations) 
The first brackets refer to the integer locations of rows, the second brackets are the selected columns 

```python
multi_rows_columns = df.iloc[[5, 7], [1, 3, 8]]
multi_rows_columns
```

---

## Using loc instead of iloc (i.e. name/index name of row or column)
Note that the column heading is typically the loc whilst the row integer (just as with iloc) will be the row loc (unless you have added an index column with names)

```python
multi_rows_columns_loc = df.loc[[5, 7], ['EdLevel', 'YearsCode']]
multi_rows_columns_loc
```

### Or a single cell.....

```python
single_cell_value = df.loc[5, 'YearsCode']
single_cell_value
```

### Or use 'slicing' to pull out a subset of data
Note that unlike normal python list slicing, the end range is included 

```python
subset_cells = df.loc[4:18, 'YearsCode']
subset_cells
```

....likewise with column slicing

```python
subset_cells = df.loc[7:9, 'YearsCode':'DevType']
subset_cells
```
